```python
import trimesh
import numpy as np
from ogsom.ogmanager import OGManager
from ogsom.mesh import MeshFace
from pathlib import Path
from ogsom.contact import Contact
from ogsom.display import line_to_mesh
```

```python
obj_path = Path('../data/cat.obj').resolve()
mesh = MeshFace.from_file(obj_path, scale=0.001)
```

```python
scene = trimesh.Scene()
scene.add_geometry(mesh.trimesh_obj)
scene.show()
```

```python
vmesh = mesh.convex_decomposition()
```

```python
scene = trimesh.Scene()
scene.add_geometry(vmesh.trimesh_obj)
scene.show()
```

```python
vmesh.faces[117].sample(0.005)
```

```python
c0 = Contact.from_facepoint(vmesh, vmesh.faces[17].sample(0.005)[3])
c1s = vmesh.find_other_contacts(c0)
```

```python
c1s
```

```python
c1 = OGManager.optimize_c1(vmesh, c0, c1s[0], 8)
```

```python
c0._point
```

```python
scene = trimesh.Scene()
scene.add_geometry(vmesh.trimesh_obj)
scene.add_geometry(line_to_mesh(c0.point, c1.point, width_offset=0.2))
scene.show()
```
